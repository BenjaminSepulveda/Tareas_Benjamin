```python
import math
import numpy as np
import matplotlib.pyplot as plt

def funcion(x):
    """Es el c√°lculo de la funci√≥n solicitada, no hay mucho que decir"""
    return 1 - math.cos(x) - math.sin(x) / x

def biseccion(funcion, a, b, tolerancia, max_iteraciones):
    """busca una soluci√≥n (ra√≠z) de una funci√≥n en un rango de valores. Divide repetidamente el rango por la mitad y verifica si la funci√≥n cambia de
       signo en ese punto. Contin√∫a este proceso hasta que encuentre una soluci√≥n aproximada dentro de cierta precisi√≥n o alcance un n√∫mero m√°ximo de
       intentos. Luego, informa sobre la soluci√≥n encontrada y la precisi√≥n lograda, o dice que no pudo encontrar una soluci√≥n en el rango especificado."""
    iteraciones = []
    errores = []

    p = (a + b) / 2.0
    i = 0

    while i < max_iteraciones:
        p = (a + b) / 2.0
        i += 1

        # C√°lculo del error relativo
        error = abs(p - (3 * math.pi / 2)) / abs(3 * math.pi / 2)
        errores.append(error)
        iteraciones.append(i)

        if funcion(p) == 0 or (b - a) / 2 < tolerancia:
            return p, iteraciones, errores

        if funcion(a) * funcion(p) < 0:
            a = p
        else:
            b = p

    return p, iteraciones, errores

a = -6.5
b = -6.0
tolerancia = 1e-8
max_iteraciones = 100
#Ahora le entregamos los valores a las variable/listas raiz, iteraciones, y errores.
raiz, iteraciones, errores = biseccion(funcion, a, b, tolerancia, max_iteraciones)

print("Ra√≠z encontrada:", raiz)

# Cota te√≥rica
cota_teorica = [(b - a) / (2 ** i) for i in iteraciones]

# Graficar el error relativo en funci√≥n de las iteraciones.
plt.semilogy(iteraciones, errores, label="Error relativo")
plt.semilogy(iteraciones, cota_teorica, label="Cota te√≥rica")
#Note que el gr√°fico esta en semilog para que se aprecie mejor la diferencia.
plt.xlabel("Iteraciones")
plt.ylabel("Error Relativo")
plt.legend()
plt.title("Error Relativo vs. Iteraciones en el M√©todo de Bisecci√≥n")
plt.grid()
plt.show()
```
