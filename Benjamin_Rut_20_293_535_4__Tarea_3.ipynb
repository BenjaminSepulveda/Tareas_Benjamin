Tienes una confusion con la definicion de media movil:

Sea una lista de datos, como mediciones de covid, la cual esta ordenada como $a_i$, donde $i=0,1,2,3,\dots N$, la media movil de largo $M$ (en el caso de la tarea $M=7$), se define como una lista de $N-M$ datos $\bar{a}_j$, con $j=M,M+1,M+2\dots N$. Cada elemento esta definido como:
$$
\bar{a}_j = \sum^{j}_{i=j-M}\frac{a_i}{M}
$$

Siguiendo esta definicion la tarea queda de la siguiente forma:
